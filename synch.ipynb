## Rsync: Copy resources directory from localhost to server

To copy only the resources directory from your local machine to the remote server:

```bash
rsync -avz /home/abv/public_html/resources root@abvchorus.org:/home/abv/public_html/
```

After copying as root, fix permissions so the abv user owns the files:

```bash
chown -R abv:abv /home/abv/public_html/resources
```

This ensures the abv user can read and write the files.
