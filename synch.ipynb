# sync

* abv is the source of truth for the static html pages and /resources. Before you do anything on dev, get current copies of all this isung rsync
* 

# to be done once
Just to setup the folder structure 

from localhost to server - assets
```bash
rsync -avz html /home/abv/public_html/assets  abv@abvchorus.org:/home/abv/public_html/
```

## on dev, stop tracking
```bash
git rm --cached public_html/*.html
git rm --cached -r public_html/resources/
git rm --cached -r public_html/assets/

```


## rsync abv -> vivo
Do this before you start working on dev

```bash
rsync -avz abv@abvchorus.org:/home/abv/public_html/*.html /home/abv/public_html/

# resources: abv->dev Delete on vivo if not on abv
rsync -avz --delete abv@abvchorus.org:/home/abv/public_html/resources /home/abv/public_html/

```
