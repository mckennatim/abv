````python
# Updated sync_with_legacy function for app.py
@app.route('/api/sync', methods=['POST'])
def sync_with_legacy():
    """Sync database with legacy HTML site"""
    try:
        logger.info("Starting sync with legacy site...")
        
        # Step 1: Download, process, and recreate songs.db from legacy site
        logger.info("Downloading and processing legacy HTML...")
        abv.create_songdb_from_legacy()
        
        # Step 2: Create new HTML file with updated content
        logger.info("Converting markdown to HTML...")
        abv.md_to_html('mdList.md', 'mdList.html')
        
        # Step 3: Recombine with header and footer
        logger.info("Recombining HTML with header and footer...")
        abv.recombine_html('header.html', 'mdList.html', 'footer.html', 'music-complete_new.html')
        
        # Step 4: Upload back to legacy site (optional)
        # abv.upload_abv(['music-complete_new.html'])
        
        logger.info("Sync completed successfully!")
        return jsonify({
            'success': True,
            'message': 'Database synced and HTML regenerated successfully',
            'steps_completed': [
                'Downloaded legacy HTML',
                'Processed and cleaned HTML',
                'Rebuilt songs.db',
                'Generated new HTML file'
            ]
        })
        
    except Exception as e:
        logger.error(f"Sync failed: {e}")
        return jsonify({
            'success': False,
            'error': str(e),
            'message': 'Sync process failed'
        }), 500
````
